```python
import pandas as pd
import glob
```

```python
path = './datasets/fundamental_data'
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

data = pd.concat(li, axis=0, ignore_index=True)
```

```python
data['growth'] = ((data['nextClossingVal']/data['clossingVal'])*100)-100
```

```python
data = data.drop(columns=['Unnamed: 0', 'fiscalDateEnding', 'reportedCurrency_x', 'nextClossingVal', 'clossingVal'])
```

```python
data = data.replace('None', 0).fillna(0)
```

```python
data = data[data['growth'] != 0]
```
